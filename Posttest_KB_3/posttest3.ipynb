```python
df = pd.read_csv("bank.csv")
fig = plt.figure(figsize = (15,8))
ax = fig.add_axes([0,0,1,1])
plt.title("Grafik Garis Antara marital dan Age")
ax.set_xlabel("Marital")
ax.set_ylabel("Age")
ax.plot(df.marital[0:10],df.age[0:10], color='purple')
plt.grid(True)
```

```python
df.to_csv("status.csv")
```

# PIE CHART

```python
df.head(20)
data_pie = df['month'].value_counts().rename_axis('month').reset_index(name='Id_count')
plt.figure(figsize=(30,26))
colors = sns.color_palette('Set2', 8)
plt.pie(data_pie.Id_count, labels=data_pie.month, colors=colors, startangle=90, autopct='%.1f%%')
plt.title('MONTH')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

#### Berdasarkan grafik pie chart di atas, terdapat 3,1% karyawan yang diterima di bulan januari, 7,0% karyawan yang diterima di bulan februari, 2,5% karyawan yang diterima di bulan maret, dan seterusnya. Data karyawan yang paling banyak diterima ada di bulan mei yaitu dengan 25,3% sedangkan paling sedikit di bulan desember.

```python
df.to_csv("month.csv")
```
