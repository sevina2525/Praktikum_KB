## Data Cleaning

```python
from sklearn.impute import SimpleImputer
```

```python
df.isna().sum()
```

## Mengubah nilai null yang punya tipe data desimal

```python
imputer = SimpleImputer(strategy = 'mean')
df["day"] = imputer.fit_transform(df[["balance"]])
print("\nJumlah null setelah menggunakan SimpleImputer: ")
print(df.isna().sum())
```

## Mengubah nilai null yang punya tipe data kategorikal

```python
imputer = SimpleImputer(strategy = 'most_frequent')
df["default"] = imputer.fit_transform(df[["default"]])
print("\nJumlah null setelah menggunakan SimpleImputer: ")
print(df.isna().sum())
```

## Mengubah nilai null yang punya tipe data bilang bulat

```python
imputer = SimpleImputer(strategy = 'most_frequent')
df["age"] = imputer.fit_transform(df[["age"]])
print("\nJumlah null setelah menggunakan SimpleImputer: ")
print(df.isna().sum())
```
